## Modifying improperly formatted data

- Fixing the data format for all the columns

```python
#Printing unique values for a list of columns in the df_Business dataframe
cols=df_Business.columns
for col in cols:
    print(col)
    
    
    uniquee = df_Business[col].unique()
    print('#.of unique values: ', len(uniquee), '\nList of unique values: ',uniquee, '\n*************************************\n')
```

```python
#Replacing values
# F-> Female, Fe male-> Female, M-> Male
df_Business['gender']= df_Business['gender'].replace(['F','Fe male','M'],['Female','Female','Male'])

#
df_Business['customer_type']= df_Business['customer_type'].replace(['Loyal','disloyal'],['Loyal Customer','Disloyal Customer'])
```

```python
#Change ease_of_online_booking unknown values to mode and 0 to 1 since 1 is the least
mode_ease = df_Business['ease_of_online_booking'].mode()
df_Business['ease_of_online_booking'] = df_Business['ease_of_online_booking'].replace(['0','Unknown'],['1', mode_ease])
```

```python
#Change inflight_wifi_service unknown values to mode and 0 to 1 since 1 is the least
mode_inflight = df_Business['inflight_wifi_service'].mode()
df_Business['inflight_wifi_service'] = df_Business['inflight_wifi_service'].replace([0,'Unknown'],[1, mode_inflight])
```

```python
#Change food_and_drink unknown values to mode and 0 to 1 since 1 is the least
mode_food_and_drink = df_Business['food_and_drink'].mode()
df_Business['food_and_drink'] = df_Business['food_and_drink'].replace(['0','Unknown','Great'],['1', mode_food_and_drink,'5'])
```

```python
#Change departure/arrival_time_convenient unknown values to mode and 0 to 1 since 1 is the least
mode_depart_arrival = df_Business['departure/arrival_time_convenient'].mode()
df_Business['departure/arrival_time_convenient'] = df_Business['departure/arrival_time_convenient'].replace([0,'Unknown'],[1, mode_depart_arrival])
```

```python
#Change inflight_entertainment unknown values to mode and 0 to 1 since 1 is the least
mode_inflight = df_Business['inflight_entertainment'].mode()
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].replace([0,'Unknown',10],[1, mode_inflight,5])
```

```python
#Change cleanliness unknown values to mode and 0 to 1 since 1 is the least
mode_cleanliness = df_Business['cleanliness'].mode()
df_Business['cleanliness'] = df_Business['cleanliness'].replace([0,'Unknown'],[1, mode_cleanliness])
```

```python
#Printing unique values for a list of columns in the df_Business dataframe
for col in cols:
    print(col)
    
    
    unique = df_Business[col].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************************************\n')
```

```python
#Replacing values in each column 


df_Business['type_of_travel']=df_Business['type_of_travel'].replace(['Business','Business travel'],['Business Travel','Business Travel'])
df_Business['baggage_handling']=df_Business['baggage_handling'].replace(['A'],['5'])
df_Business['arrival_delay_in_minutes']=df_Business['arrival_delay_in_minutes'].replace(['Three','9..'],['3','9'])
df_Business['satisfaction']=df_Business['satisfaction'].replace(['dissatisfied','0','1','satisfied','neutral'],['Dissatisfied','Neutral','Satisfied','Satisfied','Neutral'])
```

```python
#Printing unique values for a list of columns in the df_Business dataframe
for col in cols:
    print(col)
    
    
    unique = df_Business[col].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************************************\n')
```

# Handle missing values
---
Syntax:

#### DataFrame.dropna(axis=0, how='any',  inplace=False)

Arguments:
###### axis: Default – 0
   -  0, or ‘index’ : Drop rows which contain NaN values.
   - 1, or ‘columns’ : Drop columns which contain NaN value.
    
###### how: Default – ‘any’
   - ‘any’ : Drop rows / columns which contain any NaN values.
   - ‘all’ : Drop rows / columns which contain all NaN values.
    
###### inplace (bool): Default- False
   - If True, modifies the calling dataframe object

```python
#Number of missing values in each column
df_Business.isnull().sum()
```
